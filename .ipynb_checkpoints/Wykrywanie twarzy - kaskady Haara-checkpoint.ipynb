Metoda kaskad Haara zamiast analizować i obliczać dla obrazów ogromne ilości cech, korzysta się z kaskad cech. Jesli obszar/zdjęcie nie przejdzie przez 1 kaskadę nie jest już później analizowane przez kolejne, pozwalato na szybki odsiew obrazów nie będących twarzami. Można powiedzieć, że zdjęcie jest poddawane kolejnym etapom sprawdzenia, w ten sposób można odrzucić zdjęcia nie zawierające szukanego elementu(tu twarz i oczy) i nie wykonywać zbędnych obliczeń.

```python
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt
plt.rcParams["figure.figsize"] = (50,20)

face_cascade = cv.CascadeClassifier('haarcascade_frontalface_default.xml')
eye_cascade = cv.CascadeClassifier('haarcascade_eye.xml')
img = cv.imread('korwin.jpg')
gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)

faces = face_cascade.detectMultiScale(gray, 1.3, 5)
for (x,y,w,h) in faces:
    cv.rectangle(img,(x,y),(x+w,y+h),(255,0,0),2)
    roi_gray = gray[y:y+h, x:x+w]
    roi_color = img[y:y+h, x:x+w]
    eyes = eye_cascade.detectMultiScale(roi_gray)
    for (ex,ey,ew,eh) in eyes:
        cv.rectangle(roi_color,(ex,ey),(ex+ew,ey+eh),(0,255,0),2)
        
b,g,r = cv.split(img)       # get b,g,r
img = cv.merge([r,g,b])

plt.imshow(img)
plt.show()
```
